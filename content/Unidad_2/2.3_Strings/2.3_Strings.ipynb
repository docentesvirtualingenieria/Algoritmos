# <font color='157699'> **6. Indexación** </font>

Se refiere a la organización de los caracteres dentro de la cadena, y se realiza usando la posición respectiva de éstos, la cual **simpre comienza en cero**.<br>
Se usan corchetes (`[]`) para indicar los índices de una cadena de caracteres (es igual para la mayoría de estructuras de datos en Python).<br>
Si `i` es el índice de interés, se puede acceder a él mediante la siguiente sintaxis: `variable[i]`

### <font color='46B8A9'> **Indexación convencional:** </font>

Se trata de la indexación acabada de mencionar, inicia en cero $0$ para el primer elemento, y va hasta $n-1$, donde $n$ es la cantidad de elementos en la estructura de datos (`len(string)`).

La ejecucuón del siguiente código mostrará en pantalla el nombre ingresado y los indices correspondientes a cada caracter:

````python
print('Elemento','Índice', sep='\t')
print('-'*30)
for i in range(len(nombre_completo)):
    print(nombre_completo[i], i, sep='\t\t')
````

### <font color='46B8A9'> **Indexación inversa:** </font>

En Python se pueden usar indices negativos para indexar una estructura de datos. El último elemento tendrá el índice -1. Los elementos atrás de éste disminuyen su índice en -1 por cada posición:

````python
nombre_completo
````

````python
nombre_completo[-1]
````

````python
nombre_completo[-4]
````

````python
nombre_completo[3]
````

El primer elemento de cualquier estructura se puede hallar usando indexación inversa y la longitud de ésta:

````python
nombre_completo[-len(nombre_completo)]
````

La ejecucuón del siguiente código mostrará en pantalla el nombre ingresado y los indices correspondientes a cada caracter en la indexación simple e inversa:

````python
n = len(nombre_completo)
print('Elemento\t','Índice convencional\t', 'Índice invertido', sep='')
print('-'*56)
for i in range(len(nombre_completo)):
    print(nombre_completo[i], i, i-n, sep='\t\t\t')
````

# <font color='157699'> **7. Indexación por intervalos: slicing** </font>

- La indexación se puede hacer por grupos, para extraer partes de una cadena de caracteres. Esto se conoce como "*slicing*", que en inglés significa rebanado.
- La sintaxis requiere del uso de los dos puntos (`:`) para separar el inicio, el final, y el incremento en los índices:
`variable[i_ini:i_fin:incr]`
- El incremento siempre debe ser un número entero.
- Si el incremento es la unidad, se puede omitir, y usar únicamente inicio y fin: `variable[i_ini:i_fin]`
- Si se desea iniciar desde el primer carácter, no hace falta escribirlo: `variable[:i_fin]`
- Cuando se desea indexar hasta el último carácter, tampoco hace falta escribirlo: `variable[i_ini:]`

La indexación por intervalos o "*slicing*" es similar a la función `range()`, ya que al ejecutar la acción se incluye el valor de inicio, pero no el valor final. Es decir, que se llega hasta un incremento antes del valor final (si no se especifica el incremento, éste es 1 de forma predeterminada).

````python
nombre_completo
````

````python
nombre_completo[1:5]
````

````python
nombre_completo[:3]
````

````python
nombre_completo[3:]
````

````python
nombre_completo[2:5]
````

La indexación por intervalos también funciona con índices negativos:

````python
nombre_completo[-3:]
````

````python
nombre_completo[-5:]
````

Cuando no se incluye índice de inicio ni de final, se toma la cadena de caracteres completa:

````python
nombre_completo[:]
````

Se puede usar un incremento (**siempre entero**) para acceder a los caracteres "a saltos":

````python
nombre_completo[::2]
````

````python
nombre_completo[1::2]
````

Usando un incremento negativo, se puede indexar en dirección de izquierda a derecha. Por ejemplo, esto se puede usar para invertir una cadena de caracteres:

````python
nombre_completo[::-1]
````

Esta funcionalidad permite reconocer palíndromos de forma sencilla en Python:

````python
palindromo = 'reconocer'
palindromo == palindromo[::-1]
````

**Resumen:** <br> sea `s = 'abcdefghij'` <br>

| Código | Resultado | Descripción|
| :--- | :--- | :--- |
|`s[2:5]` |`cde` | Caracteres en los índices 2 a 4 |
|`s[:5]` |`abcde` | Primeros cinco caracteres (0 a 4) |
|`s[5:]` | `fghij` | Caracteres desde el índice 5 hasta el final |
|`s[-2:]` | `ij` | Dos últimos caracteres |
|`s[:]` | `abcdefghij` | Cadena de caracteres completa |
|`s[1:7:2]` | `bdf` | Caracteres del índice 1 al 6, cada dos posiciones |
|`s[::-1]` | `jihgfedcba` | Un paso negativo invierte la cadena de caracteres |

# <font color='157699'>**9. Conversión a número y viceversa**</font>

Las cadenas de caracteres que contienen únicamente números, o los signos  `+ ` o `-` al inicio, se pueden covertir en números haciendo uso de la función conversión a enteros: `int()`

Las cadenas de caracteres que tienen la forma de un número con cifras decimales, para lo cual se hace uso del seperador decimal punto: `.` se pueden convertir en números reales usando la función: `float()`.

De igual forma, valores numéricos que requieran expresarse como números complejos, usando para ello la representación convencional de la unidad imaginaria como la letra `j`, se pueden convertir usando la función `complex()`.

Observe que para poder operar numéricamente una cadena de caracteres que tiene la forma de un número primero debe hacer la conversión de formato, por esta razón la ejecución de las siguientes instrucciones generarán errores:

```
Nombre = '286'
Numero = 1350 + 150
Nombre + 14
complex(Nombre) + 14
```

````python
float(Nombre) + 14
````

Conversión a complejos: `complex()`

# <font color='157699'>**11. Métodos adicionales para cadenas de caracteres**</font>


En Python, las cadenas de caracteres (y en general los diferentes tipos de variables) tienen asociados métodos que pueden aplicarse de forma directa, ahorrando código. El concepto de método hace parte de la programación orientada a objetos (POO), la cual se revisará brevemente al final de este curso. Sin embargo, de momento puede pensar en estos métodos como acciones que se pueden aplicar a la variable (llamada objeto). La sintaxis para usar estos métodos es:

> `nombre_variable.método()`



Los métodos `lower()` y `upper()` convierten una cadena de carcteres completa a minúsculas y mayúsculas, respectivamente.

````python
'PRUEBA'.lower()
````

````python
palabra = 'ESTATUTO'
````

````python
palabra.lower()
````

Estos método no cambian el valor de la variable, sólo la transforman temporalmente:

````python
palabra
````

Para cambiar una variable con un método de este tipo, se debe sobreescribir su valor:

````python
palabra = palabra.lower()
palabra
````

````python
palabra.upper()
````

````python
palabra
````

Este método reconoce únicamente letras, no números ni carcteres especiales, ni el espacio.

````python
palabra.isalpha()
````

````python
Nombre = '286'
Nombre
````

````python
Nombre.isalpha()
````

Del mismo modo, existe un método que permite verificar si todos los caracteres son números:

````python
Nombre.isalnum()
````

Este método tampoco cambia el valor original de la variable:

````python
palabra
````
