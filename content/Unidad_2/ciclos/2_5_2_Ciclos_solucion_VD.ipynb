Los algoritmos y los programas que los implementan usualmente deben ejecutar una instrucción o bloques de instrucciones repetidamente. Las estructuras que repiten una secuencia de instrucciones un número determinado de veces se denominan bucles y

### <font color='46B8A9'> **Ejemplo 1.** </font>
Usted desea estimar el promedio de sus notas este semestre. La siguiente lista contiene sus calificaciones en las seis asignaturas que curso:
 ```
notas = [3.5, 4.5, 3.8, 4.9, 3.7, 4.3]
```
¿Cómo obtendría usted manualmente el promedio de las notas? 
Considere el siguiente programa para calcular el promedio



````python
notas = [3.5, 4.5, 3.8, 4.9, 3.7, 4.3]
total = 0 # la suma inicial es 0

for i in range(len(notas)): # Repite para todas las notas que tenga
    total = total + notas[i] # Suma la nota a total

promedio = total/len(notas) # Divide el total de todas las notas entre el número de asignaturas
print("Su promedio es: ", promedio)# Imprime promedio
````

### <font color='46B8A9'> **Ejemplo 2:** </font> 
El programa genera un número aleatorio entre 0 y 9 que usted desconoce, usted deberá seleccionar números hasta que logre adivinarlo.

````python
import random # importa random

aleatorio = random.randint(0, 9) # Genera un número aletaorio entre 0 y 9
noAdivina = True # no Adivina el número

while noAdivina:
    escogidoStr = input("Ingrese un número entre 0 y 9 ") # Ingresar número
    escogidoInt = int(escogidoStr) # Convierte a entero el número ingresado como string

    # Verifica si adivina el número
    if aleatorio == escogidoInt : 
        print("Has ganado! Felicitaciones!")
        noAdivina = False # adivina el número y cambia a False 
    else: 
        print("No Has ganado, sigue intentado")
````

Note que en este caso no se conoce de antemano el número de veces que debe repetirse la instrucción que solicita ingresar un número. 




## <font color='157699'> **Ciclo `while`** </font>

Esta estructura se encarga de ejecutar una misma instrucción o conjunto de instrucciones "mientras que" una determinada condición se cumpla. Ejemplo: Mientras que el número ingresado no sea igual al generado por el computador, solicite un nuevo número

El cuerpo del bucle se repite mientras se cumple una determinada condición. Al ejecutar la estructura, primero se evalúa la condición (una expresión booleana). Si la expresión se evalúa falsa, no se toma ninguna acción y el programa prosigue en la siguiente instrucción despues del bucle o ciclo `while`. Si la expresión booleana es verdadera, entonces se ejecutan las instrucciones en el cuerpo del while, después de lo cual se evalúa de nuevo la expresión booleana. Este proceso se repite una y otra vez mientras la expresión booleana (condición) sea verdadera

![title](https://docs.google.com/uc?export=download&id=1DbruR6MCOSkxgVIbjxURHWAV0sqQ7bab)



La sintaxis para escribir este tipo de estructuras cambia dependiendo del lenguaje pero en general podrá verse de la siguiente forma:
```
while (Condición) do
    Instrucciones a realizar si la condición es verdadera
end while
```

Retomemos el ejemplo del inicio de la sección. 

> ¿Cuál es la condición de parada del ciclo? ¿Qué instrucciones repite cada vez que ejecuta una iteración del ciclo?

````python
import random # importa random

aleatorio = random.randint(0, 9) # Genera un número aletaorio entre 0 y 9
noAdivina = True # no Adivina el número

while noAdivina:
    escogidoStr = input("Ingrese un número entre 0 y 9 ") # Ingresar número
    escogidoInt = int(escogidoStr) # Convierte a entero el número ingresado como string

    # Verifica si adivina el número
    if aleatorio == escogidoInt : 
        print("Has ganado! Felicitaciones!")
        noAdivina = False # adivina el número y cambia a False 
    else: 
        print("No Has ganado, sigue intentado")
````

## <font color='157699'> **Ciclo `for`** </font>

En muchas ocasiones se conoce de antemano el número de veces que se desean ejecutar las acciones de un bucle o ciclo.
En estos casos, en el que el número de iteraciones es fijo, se debe usar la estructura `for`. Esta estructura ejecuta las acciones del cuerpo del bucle un número especificado de veces y de modo automático

![title](https://docs.google.com/uc?export=download&id=1tR7wZzh-kMWaKulc40o6QxlukyAGpunc)



La sintaxis para escribir este tipo de estructuras cambia dependiendo del lenguaje pero en general podrá verse de la siguiente forma:
```
for i = a hasta b do
    Instrucciones a realizar si la condición es verdadera
end for
```
