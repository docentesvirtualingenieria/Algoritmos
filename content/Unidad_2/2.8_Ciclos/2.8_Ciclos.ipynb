##</font> <font color='8EC044'> **Estructuras iterativas** </font>

````python
notas = [3.5, 4.5, 3.8, 4.9, 3.7, 4.3]
total = 0 # la suma inicial es 0

for i in range(len(notas)): # Repite para todas las notas que tenga
    total = total + notas[i] # Suma la nota a total

promedio = total/len(notas) # Divide el total de todas las notas entre el número de asignaturas
print("Su promedio es: ", promedio)# Imprime promedio
````

## <font color='157699'> **Contadores** </font>


La construcción de un contador es una de las técnicas más comunes en la realización de algoritmos.
Es una variable en la memoria que se incrementará en una unidad cada vez que se ejecute el proceso.
El contador se utiliza para llevar la cuenta de determinadas acciones que se pueden solicitar durante la resolución de un problema.
En las instrucciones de preparación se realiza la inicialización del contador o contadores. La inicialización consiste en poner el valor inicial de la variable que representa al contador.
Generalmente se inicializa con el valor 0 para operaciones de suma y en 1 para operaciones de multiplicación.


## <font color='157699'> **Acumuladores** </font>


Un acumulador  es una variable en la memoria cuya misión es almacenar cantidades variables.
Se utiliza para efectuar sumas sucesivas. La principal diferencia con el contador es que el incremento o decremento de cada suma es variable en lugar de constante como en el caso del contador.

## <font color='157699'> **Contadores y Acumuladores** </font>


El ejemplo más sencillo para explicar contadores y acumuladores detalla el promedio de n números ingresados a un sistema.
Un contador debe llevar cuenta de la cantidad de números ingresados y el acumulador debe llevar la suma de todos los valores ingresados, al final del procedimiento realizamos la siguiente operación,
<h1>
${promedio} = \frac{acumulador}{contador}$


## <font color='157699'> **La función `range`** </font>

Si quisiéramos iterar a través de los valores de 0 a 4, podríamos hacer esto:


```python
for n in (0, 1, 2, 3, 4):
    print(n)
```
Esta solución no es tan mala cuando solo hay unos pocos números. Pero si el rango de números fuera mucho mayor, rápidamente se volvería tedioso.

Python ofrece una mejor opción: la función incorporada `range()`, que devuelve un iterable que produce una secuencia de números enteros.

`range(<end>)` devuelve un iterable que produce números enteros que comienzan con `0`, hasta pero sin incluir `<end>`:


````python
x = range(5)
x
````

````python
print(type(x))
````

Tenga en cuenta que `range()` devuelve un objeto de clase range, no una lista o tupla de los valores. Debido a que un objeto de rango es iterable, puede obtener los valores iterando sobre ellos con un ciclo `for`:

```python
for n in x:
    print(n)
```
También puede relacionar todos los valores a la vez con `list()` o `tuple()`. En una sesión [REPL](https://en.wikipedia.org/wiki/Read%E2%80%93eval%E2%80%93print_loop), esa puede ser una forma conveniente de mostrar rápidamente cuáles son los valores:


````python
list(x)
````

````python
tuple(x)
````

Sin embargo, cuando se usa `range()` en código que es parte de una aplicación más grande, generalmente se considera una mala práctica usar `list()` o `tuple()` de esta manera. Los valores en el rango especificado no se generan hasta que se solicitan. El uso de `list()` o `tuple()` en un objeto `range` obliga a que todos los valores se devuelvan a la vez. Esto rara vez es necesario y, si la lista es larga, se pierde tiempo y memoria.

Es muy importante tener claro que la funcion `range()` no toma el valor final.

`range(<inicio>, <fin>, <paso>)` devuelve un iterable que produce enteros que comienzan con `<inicio>`, hasta pero sin incluir `<fin>`. Si se especifica, `<paso>` indica una cantidad para omitir entre valores:


```python
list(range(1, 10, 2))
>>>[1, 3, 5, 7, 9]
```

Si se omite <paso>, el valor predeterminado es 1:
```python
list(range(5, 10, 1))
>>>[5, 6, 7, 8, 9]
list(range(5, 10))
>>>[5, 6, 7, 8, 9]
```
Todos los parámetros especificados en `range()` deben ser números enteros, pero cualquiera de ellos puede ser negativo. Naturalmente, si <inicio> es mayor que <fin>, <paso> debe ser negativo si quieres algún resultado:



```python
list(range(-5, 5))
>>> [-5, -4, -3, -2, -1, 0, 1, 2, 3, 4]

list(range(5, -5))
>>> []

list(range(5, -5, -1))
>>> [5, 4, 3, 2, 1, 0, -1, -2, -3, -4]
```

## <font color='157699'> **Ciclo `while`** </font>

Esta estructura se encarga de ejecutar una misma instrucción o conjunto de instrucciones "mientras que" una determinada condición se cumpla. Ejemplo: Mientras que el número ingresado no sea igual al generado por el computador, solicite un nuevo número

El cuerpo del bucle se repite mientras se cumple una determinada condición. Al ejecutar la estructura, primero se evalúa la condición (una expresión booleana). Si la expresión se evalúa falsa, no se toma ninguna acción y el programa prosigue en la siguiente instrucción después del bucle o ciclo while. Si la expresión booleana es verdadera, entonces se ejecutan las instrucciones en el cuerpo del while, después de lo cual se evalúa de nuevo la expresión booleana. Este proceso se repite una y otra vez mientras la expresión booleana (condición) sea verdadera


![title](https://docs.google.com/uc?export=download&id=1DbruR6MCOSkxgVIbjxURHWAV0sqQ7bab)



La sintaxis para escribir este tipo de estructuras cambia dependiendo del lenguaje pero en general podrá verse de la siguiente forma:
```python
while (Condición):
    Instrucciones a realizar si la condición es verdadera
```

Retomemos el ejemplo del inicio de la sección.

> ¿Cuál es la condición de parada del ciclo? ¿Qué instrucciones repite cada vez que ejecuta una iteración del ciclo?

````python
import random # importa random

aleatorio = random.randint(0, 9) # Genera un número aletaorio entre 0 y 9
noAdivina = True # no Adivina el número

while noAdivina:
    escogidoStr = input("Ingrese un número entre 0 y 9 ") # Ingresar número
    escogidoInt = int(escogidoStr) # Convierte a entero el número ingresado como string

    # Verifica si adivina el número
    if aleatorio == escogidoInt :
        print("Has ganado! Felicitaciones!")
        noAdivina = False # adivina el número y cambia a False
    else:
        print("No Has ganado, sigue intentado")
````

**Ejemplo: Identificación del tipo de elementos en una lista**

````python
lista = ['o', 5, 7.2, 'palabra', 10]
ent = 0 # cantidad de enteros
cad = 0 # cantidad de cadenas de caracteres
real = 0 # cantidad de números reales
for item in lista:
    if type(item) == int:
        ent += 1
    elif type(item) == str:
        cad += 1
    elif type(item) == float:
        real += 1
print('\nLa lista tiene %d elementos que son números enteros' % ent)
print('\nLa lista tiene %d elementos que son números reales' % real)
print('\nLa lista tiene %d elementos que son cadenas de caracteres' % cad)
````

Las listas también se pueden recorrer haciendo referencia explícita al índice del elemento. En este caso es necesario indicar la secuencia de indices, para lo que usamos la función `len()`, con la cual se obtiene la cantidad de elementos, así como la función `range()`, con al cual se crea la secuencia de índices.

````python
lista = [1, 2, "a", 6]
for i in range(len(lista)):
    print(lista[i])
````

##<font color='157699'>**Recorrido o iteración sobre diccionarios**</font>

Los elementos de un diccionario se pueden recorrer de distintas formas, dependiendo de si accedemos a las claves, a los valores, o a las parejas (clave : valor).


````python
dic
````


Podemos recorrer directamente los valores de un diccionario usando el método `values()`



````python
for valor in dic.values():
    print(valor)
````

De igual forma podemos iterar sobre los valores, conociendo las claves, para ello hacemos uso del método `keys()`

````python
for key in dic.keys():
    print(dic[key])
````

````python
for clave, valor in dic.items():
    print("La clave", clave, 'tiene asociado el valor', valor)
````

##<font color='157699'>**Programación compacta con diccionarios <font color='46b8a9'> (*dict comprehensions*)**</font>

La programación compacta con diccionarios (`dictionary comprehension`) es una construcción de Python que permite crear un diccionario utilizando una sintaxis compacta y legible. Es similar a la programación compacta con listas y con tuplas, pero en lugar de crear una lista, se crea un diccionario.

La sintaxis de la programación compacta con diccionarios es la siguiente:

> ```python
    {clave:valor for variable in iterable}

Donde `clave` es la clave que se le asigna al valor `valor`, `variable` es la variable utilizada en la iteración, e `iterable` es el objeto que se itera (como una lista o una tupla).

````python
lista = ['Algoritmos', 'Programación', 'Simulación']
dic = {palabra:len(palabra) for palabra in lista}
dic
````

También se pueden agregar condiciones a la programación compacta con diccionarios, mediante la siguiente sintaxis:

Consiste en definir una o varias operaciones, incluidas la evaluación de condiciones y la estructuras de repetición, directamente en una línea de código entre corchetes:
    
```python
[f(var) for var in iterable if var > 0]
```
    
Donde `f()` es una función que genera el valor de cada elemento de la lista. `var` es una variable transitoria que toma los valores de los elementos en la estructura de datos `iterable`, si se cumple la condición dada.
    
La programación compacta con listas ahorra mucho código, y permite crear resultados útiles de forma rápida y sencilla.
    
Por ejemplo, dada una lista de elementos `L`, estamos interesados en crear una nueva lista con cada uno de sus elementos elevados al cuadrado, sólo si éstos son mayores que 4.

La forma convecional, usando ciclos `for`, sería:  

````python
lista1 = [1, 2, 3, 4, 5, 6]
lista2 = []
for item in lista1:
    if item > 4:
        lista2.append(item**2)
lista2
````

La forma compacta usado `list comprehensions` sería (Se usa una línea de código, en vez de 4):

````python
lista1 = [1, 2, 3, 4, 5, 6]
lista2 = [item**2 for item in lista1 if item > 4]
lista2
````

###<font color='46B8A9'> **Ejemplo 9**</font>

Listas de números aleatorios

La función `randint(a, b)` de la libreria `random`, genera un número entero entre `a` y `b`, incluyendo ambos extremos.

````python
from random import randint
randint(5,24)
````

Usemos esta función para crear una lista de 15 números aleatorios entre 1 y 10, a traves de la programación compacta con listas.

````python
lista = [randint(1, 10) for i in range(15)]
lista
````

Ahora podemos usar nuevamente la programación compacta con listas para crear el conteo de las frecuencias de cada valor entre 1 y 10 en la lista creada previamente.

````python
frecuencias = [lista.count(i) for i in range(1,11)]
frecuencias
````

Esta información se puede mostrar de forma organizada:

````python
print('Valor', 'Repeticiones', sep='\t')
print('-'*20)
for i in range(10):
    print(i+1, frecuencias[i], sep='\t')
````
