# <font color='056938'> **Variables** </font>

Una *Variable* puede ser entendida como un contenedor de datos, en t√©rminos de computaci√≥n es un espacio reservado en la memoria del computador para almacenar un dato que puede ser usado o modificado tantas veces como se desee. El buen uso de la variables parte del uso de nombres apropiados

* Tener un nombre que sea f√°cilmente interpretable en el c√≥digo
* No tener el mismo nombre que una ‚Äúpalabra reservada‚Äù del lenguaje. <a href="https://es.wikibooks.org/wiki/Python/Generalidades/Palabras_reservadas,_operadores_y_s%C3%ADmbolos_del_lenguaje">Palabras reservadas en Python</a>
* S√≥lo pueden ser letras, d√≠gitos y el guion bajo o subguion.
* Deben comenzar por una car√°cter (letra).
* No puede incluir espacios en blanco.
* No pueden llevar acento (tilde)


Consideremos el ejemplo del juego con el que iniciamos la secci√≥n. Corre el siguiente c√≥digo hasta que ganes:



````python
import random # importa la libreria random

nombre = input("¬øCu√°l es su nombre?:    ") # Obtener el nombre del usuario
aleatorio = random.randint(1, 10) # Genera un n√∫mero aletaorio entero entre 1 y 10
jugador = int(input("Ingrese un n√∫mero entero entre 1 y 10: ")) # Pregunta un n√∫mero al usuario

print(f'\nPC gener√≥ el n√∫mero n√∫mero: {aleatorio}')
print(f'{nombre}, ingresaste el n√∫mero: {jugador}')

print(f'\n{nombre} ganaste! üôÇ' if jugador>aleatorio else '{nombre} perdiste! üôÅ') # M√°s adelante conoceras en detalle como funciona la instrucci√≥n if y else
````

## <font color='157699'> **Tipos de datos** </font>

Los datos almacenados en una variable pueden ser de diversos tipos. El tipo de datos es el atributo que especifica al computador la clase de dato que tiene que manejar, para saber qu√© valores puede tomar y qu√© operaciones realizar. Algunos de los tipos de datos b√°sicos (usualmente llamados tipos primitivos o elementales) son:

* N√∫meros enteros: ...-3, -2, -1, 0, 1, 2, 3, ...
* N√∫meros en coma flotante (decimales). Dependiendo del lenguaje podr√≠an recibir distintos nombres o clasificarse de distintas maneras como *double* o *float*
* Caracteres (alfanum√©ricos) - `char`
* Cadenas de caracteres (textos) - `str`
* Estados l√≥gicos (booleanos) - `int`

Note que existen una gran cantidad de tipos de datos y que incluso la disponibilidad de estos pueden cambiar dependiendo del lenguaje.

Veamos un ejemplo de cada uno de ellos:


````python
entero = 4 # variable entera
print("Una variable entera: ", entero)
decimal = -1.3 # variable real
print("Una variable real: ", decimal)
caracter = "a" # variable caracter
print("Una variable caracter: ", caracter)
cadena = "Pablo" # variable cadena
print("Una variable string: ", cadena)
estadoLogico = False # variable booleano
print("Una variable de estado l√≥gico: ", estadoLogico)
````

# <font color='056938'> **Operadores** </font>

Los operadores realizan operaciones con variables y pueden clasificarse segun el tipo de operaci√≥n que realizan y el tipo de variables que manipulan. En particular algunos de los tipos de operadores que distingue Python son:  

## <font color='157699'> **Operadores algebraicos** </font>
| Operador| Operaci√≥n 	| Descripci√≥n 	| Ejemplo (a=10 - b=20)|
|:---|:---|:---|:---|
| + | Suma 	| Suma los valores a ambos lados del operador 	| a + b = 30 	|
| -  | Resta 	| Resta el valor de la derecha al valor que esta a la izquierda 	| a ‚Äì b = -10 	|
| *  | Multiplicaci√≥n 	| Multiplica los valores a los dos lados del operador 	| a * b = 200 	|
| / | Divisi√≥n 	| Divide el valor de la izquierda sobre el de la derecha	| b / a = 2.0 	|
| // | Divisi√≥n entera	| Divide el valor de la izquierda sobre el de la derecha	| b // a = 2 	|
| % | M√≥dulo 	| Residuo de la divisi√≥n de dos n√∫meros 	| b % a = 0 	|
| ** | Exponente 	| Eleva a la potencia 	| a**b = $10^{20}$	|

Observe los siguientes ejemplos

````python
# Observe los siguientes ejemplos

# Dado los n√∫meros enteros a y b:
a = 21
b = 10

c = a + b
print(f"{a} + {b} = {c}")

c = a - b
print(f"{a} - {b} = {c}")

c = a * b
print(f"{a} * {b} = {c}")

c = a / b
print(f"{a} / {b} = {c}")

c = a // b
print(f"{a} // {b} = {c}")

c = a % b
print(f"{a} % {b} = {c}")

a = 2
b = 3
c = a**b
print(f"{a} ** {b} = {c}")
````

## <font color='157699'> *Jerarquia operadores algebraicos* </font>
En programaci√≥n, cuando se construye una expresi√≥n matem√°tica o f√≥rmula se debe tener en cuenta la jerarqu√≠a de operadores. La jerarqu√≠a de operadores determina el orden en el que se resuelven las expresiones cuando se involucran operaciones aritm√©ticas como la suma, resta, multiplicaci√≥n, divisi√≥n, potencia, ra√≠z y m√≥dulo de la divisi√≥n.

1. Par√©ntesis.
2. Exponenciaci√≥n, radicaci√≥n y m√≥dulo.
3. Divisi√≥n y multiplicaci√≥n.
4. Suma y resto.

* Si los operadores estan en el mismo nivel de jerarquia se realizan de izquierda a derecha

* Si la operaci√≥n contiene par√©ntesis, se realizan las operaciones que se encuentran en el par√©ntesis m√°s interno, siguiendo el nivel de jerarquia mencionado anteriormente.

## <font color='157699'> **Operadores de comparaci√≥n** </font>

Comparan los valores a ambos lados del operador y deciden la relaci√≥n entre dichos valores. Son tambi√©n conocidos como operadores relacionales. El resultado de estas operaciones son almacenados en variables booleanas, es decir, aquellas variables que toman un valor verdadero (1 o True) o falso (0 o False).


| Operador 	| Descripci√≥n 	| Sintaxis	| Ejemplo (a=10 y  b=20)|
|----------	|-----------------------------------------------------------------------------------------------------------------------	|------------------------------------------------	|-------------|
| == 	| Verdadero si los dos operandos son iguales 	| a == b| False |

| <> 	| Verdadero si los dos operandos son diferentes (usado en otros lenguajes de prog.)	| a <> b | True |

| < 	| Verdadero si el operando de la izquierda es menor que el de la derecha 	| a < b | True |
| >= 	| Verdadero si el operando de la izquierda es mayor o igual que el de la derecha | a >= b	| False |
| <= 	| Verdadero si el operando de la izquierda es menor o igual que el de la derecha 	| a <= b	| True |




Por ejemplo, observe que valor toman las variables P y Q en las siguientes celdas:

## <font color='157699'> **Operadores l√≥gicos** </font>

Los operadores l√≥gicos son elementos fundamentales en la l√≥gica y la programaci√≥n que permiten combinar o comparar expresiones l√≥gicas para obtener un resultado basado en ciertas condiciones. Estos operadores se utilizan en diferentes contextos, como en programaci√≥n, matem√°ticas, electr√≥nica y sistemas de control.

En la mayor√≠a de los lenguajes de programaci√≥n, los operadores l√≥gicos trabajan con valores booleanos. Los operadores l√≥gicos b√°sicos son tres:

| Operador 	| Descripci√≥n 	|
|-----------------	|----------------------------------------------------------------------	|
|  and 	| Verdadero si los dos operandos son verdaderos 	|
|  or 	| Verdadero si uno de los operandos es verdadero 	|
|  not 	| Reversa el estado l√≥gico de un operando 	|

Considere por ejemplo la siguiente expresi√≥n

```python
(a > b) or not(b < c)
```
En esta aparecen dos expresiones l√≥gicas `(a > b)` y `not(b < c)` unidas por el operador l√≥gico `or`. El resultado de la siguiente expresi√≥n depender√° entonces de los valores que tome cada una de las expresiones l√≥gicas y que se puede resumir como

| `(a>b)`	| `not(b<c)` | `(a > b) or not(b < c)`|
|-----------------	|-----------------------------------	|--------|
|  False 	| False	| False |
|  True 	| False	| True |
|  False 	| True | True |
|  True 	| True | True |
