# <font color='056938'> **Notebooks - </font> <font color='8EC044'> Conceptos básicos** </font>

Jupyter Notebook es una aplicación web de código abierto que nos permite crear y compartir documentos interactivos que contienen código, visualizaciones y texto explicativo. Es una herramienta muy popular entre científicos de datos, ingenieros, investigadores y educadores, ya que facilita la realización de tareas de análisis de datos, modelado, simulaciones y más.

La principal característica de Jupyter Notebook es que nos permite trabajar con celdas. Cada celda puede contener texto formateado utilizando el lenguaje de marcado Markdown o código en varios lenguajes de programación, como Python, R, Julia, entre otros. Esto significa que se puede escribir una explicación en una celda, y luego en otra celda, ejecutar código para mostrar resultados o gráficos.

Por ejemplo, la celda que actualmente estas leyendo es una celda de texto mientras que la celda que se presenta a continuación permite ejecutar el código que escribas.

LaTeX es un lenguaje de programación orientado a la generación de textos de alta calidad y comunmente usado en producción de textos y documentos científicos. A modo de ejemplo observe como pueden generarse ecuaciones con distintos formatos en latex<br><br>

* Ecuaciones en la misma línea de texto: En este caso, la ecuación se inserta usando el carácter \$, así, `$ ecuación $`. Por ejemplo para incluir la ecuación $\sqrt{\alpha^2 - 3\beta}$ en una misma linea de texto se utiliza la sintaxis `$\sqrt{\alpha^2 - 3\beta}$`. <br><br>

## <font color='157699'> **Asignación múltiple de variables** </font>
Python a diferencia de otros lenguajes de programación permite hacer una asignación múltiple. Por ejemplo, suponga que tienes dos variables `a` y `b`, para intercambiar su contenido se utiliza la siguiente sintáxis. Por ejemplo, si inicialmente `a=10` y `b=20`, el código que implementes debe hacer que el contenido se intercambie haciendo que `a=20` y `b=10`.

````python
#Variables
a = 10
b = 20

#Muestra el contenido de cada variable
print(f'a={a} y b={b}')

#Intercambia el contenido de las variables
a, b = b, a

#Muestra el contenido de cada variable
print(f'a={a} y b={b}')
````

## <font color='157699'> **Funciones predeterminadas** </font>


El interpreter de un lenguaje, en nuestro caso Python, admite un conjunto de funciones que estan prediseñadas, es decir, no es necesario importar ninguna libreria para usarlas. Por ejemplo, la función `input()` es una función predefinida. En este [link](https://docs.python.org/es/3/library/functions.html) puedes encontrar más información sobre estas funciones.

 A modo de ejemplo a continuación se presentan algunas funciones numéricas predefinidas.


| Función 	| Descrpición 	|
|----------	|---------------------------------------------------------------------	|
| abs() 	| Retorna el valor absoluto de un número 	|
| divmod() 	| Retorna una tupla con el cociente  y el residuo de la división entera 	|
| max() 	| Retorna el máximo de una colección de datos 	|
| min() 	| Retorna el mínimo de una colección de datos 	|
| pow() 	| eleva un número a una potencia dada	|
| round() 	| Redondea un número a su entero más cercano	|
| complex() | Retorna el número complejo con el valor real + imag*1j|

Considere por ejemplo:

````python
a = 5
b = 2
print("El modulo de la división entre a y b es: ", divmod(a, b))
print("El residuo de la división entera entre a y b es: ", a%b)
````



**Números complejos**

En python podemos operar con número complejos empleando la función predeterminada `complex()`. Por ejemplo, definamos `Z1` como un número complejo, usando la expresión *1j* para definir el número imaginario $\sqrt{-1}$, así

````python
Z1 = 4 - 3 * 1j
Z1
````

Podemos validar que tipo de dato es con la función `type()`

````python
type(Z1)
````

````python
Z = complex(3,4)
Z
````

````python
e = 23
````

````python
complex(e)
````

````python
Z + Z1
````

````python
Z
````

Note sin embargo, que no podemos convertir un número complejo en un entero o un flotante por lo que el siguiente fragmento de código generará error

````python
#int(Z)
````

````python
#float(Z)
````

Para obtener la parte real e imaginaria del número complejo se puede emplear los métodos `real` e `imag`, así

````python
Z = 7-4*1j
Z_real = Z.real
Z_imaginario = Z.imag
print(f'La parte real e imaginaria del número complejo {Z} es {Z_real} y {Z_imaginario}, respectivamente')
````

Ahora cada uno de las partes del número complejo pueden ser convertidas a entero o flotante para su manipulación o la construcción de otro número complejo, por ejemplo

````python
int(Z_real)
````

````python
int(Z.imag)
````

````python
Z2 = (4*Z_real,3+Z_imaginario)
Z2
````

## <font color='157699'> **Importación de librerías** </font>

En muchos casos, las funciones y métodos que se requieren utilizar no estan disponibles de forma predeterminada en el lenguaje *Python* y por lo tanto deben ser adicionadas a partir de librerias.

Las librerías son conjuntos de programas especializados para un fin específico, por ejemplo, la generación de gráficas, cálculos numéricos, procesamiento de datos e imágenes, entre otros.

Para importar las librerias se tienen las siguientes opciones:

### <font color='46B8A9'> **Opción 1:** importar con el nombre original: </font>

Por ejemplo, la librería `math` es una librería estandar de Python y proporciona acceso diversas funciones matemáticas. Para poder hacer uso de sus funciones se debe importar, así

````python
import math
````

Y para usar una función se especifica el nombre de la libreria junto con la función. Por ejemplo, si se desea usar la función raíz cuadrada (`sqrt()`) de la libreria `math` se escribe

````python
math.sqrt(25)
````

### <font color='46B8A9'> **Opción 2:** importar con un alias </font>

La librería NumPy (Numerical Python) es una de las librería fundamentales más importantes en el ecosistema de Python para la computación científica y el análisis de datos.

Para evitar tener que escribir todo el nombre de la librería en conjunto con la función se puede importar la librería con un alias. En este caso usaremos el alias `np` para referir a la librería `numpy`.

````python
import numpy as np
````

````python
np.sqrt(36)
````

En este caso ya no es posible usar el nombre de la librería para acceder a las función. Por ejemplo, si se descomenta el siguiente código se obtendría un error.

````python
#numpy.sqrt(36)
````

Usted podría usar cualquier alias. Sin embargo, es frecuente usuar alias comunes dentro de la comunidad. Por ejemplo, `np` para la librería `numpy` o `pd` para la librería `pandas`. Por ejemplo, la libreria `numpy` se puede importar con el alias `sofia` como

````python
import numpy as sofia
````

Y se usa la función anteponiendo el alias de la librería a la que pertenece, así

````python
sofia.sqrt(36)
````

### <font color='46B8A9'> **Opción 3:** Importar directamente funciones de una librería </font>

Es posible importar funciones específicas de una librería. Por ejemplo, para importar la función `sqrt` de la librería `numpy` se usa la sintaxis:

````python
from numpy import sqrt
````

Ahora es posible usarla sin usar ningún alias, así

````python
sqrt(16)
````

También puedes usar un alias para la función, en la siguiente celda se usa el álias `raiz` para la función `sqrt` que se importa

````python
from numpy import sqrt as raiz
````

Y se usa directamente la función con el alias utilizado

````python
raiz(49)
````
