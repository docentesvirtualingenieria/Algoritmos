# <font color='157699'> **Listas** </font>
Son estructuras unidimensionales, mutables (se puede cambiar el valor de sus elementos de forma independiente) y permiten anidar otras estructuras de datos.
Entre sus características se tiene:
- Estructuras de datos heterogéneas: pueden combinar tipos de datos diferentes (numéricos, cadenas de caracteres, etc.)
- Similares en muchos aspectos a las cadenas de caracteres
- Se usan frecuentemente en la programación con Python

# <font color='157699'> **1. Creación** </font>
    
Para crearlas se escriben sus elementos entre corchetes (`[]`), separándolos con coma:

````python
lista = [1, 2, 3]
lista
````

````python
type(lista)
````

### <font color='46B8A9'> **Heterogeneidad** </font>

Note que una lista puede combinar elementos de distinto tipo, es decir, hetereogéneos:

````python
lista = [1, 3.8, 5, 'carlos']
lista
````

````python
type(lista)
````

Incluso, es posible contener una lista dentro de otra lista:

````python
lista = [1, 3.8,[3, 'Juan', False], 'carlos']
lista
````

### <font color='46B8A9'> **Lista vacía** </font>  
Equivalente al cero y a la cadena de caracteres vacía. Se pueden añadir elementos posteriormente:

````python
lista = []
lista
````

````python
lista = list()
lista
````

### <font color='46B8A9'> **Ingreso de listas desde teclado con el comando `input`**  </font>

En algunas ocasiones podemos obtener los elementos de una lista escritos como una cadena de caracteres, por ejemplo cuando la recibimos por teclado, al ingresar:
> `[3, 5, 6, 7]`




El comando `eval()` interpreta la cadena de caracteres y la convierte en una lista, al identificar los corchetes.

````python
lista = input('Ingrese una lista usando corchetes: ')
print(lista, " es de tipo ", type(lista))
````

````python
lista = input('Ingrese una lista usando corchetes: ')
lista = eval(lista)
print(lista, " es de tipo ", type(lista))
````

Es posible concatenar directamente los corchetes para ahorrar una línea de código:

````python
lista = eval('[' + input('Ingrese una secuenca de elementos separado por comas: ') + ']')
print(lista, " es de tipo ", type(lista))
````

### <font color='46B8A9'> **Comando `list()`** </font>  


Esta función permite crear una lista a partir de un argumento, siempre y cuando el tipo de datos del argumento permita su conversión a una lista:

Por ejemplo, la instrucción `range(1,10)` crea una secuencia de números entre 1 y 9, estos pueden ser almacenados en una lista

````python
rango = range(1,10)
lista = list(rango)
print(rango) # imprime el rango
print(lista) # imprime la lista
````

````python
type(lista)
````

Al pasar como argumento una cadena de caracateres, obtenemos una lista de con cada uno de los caracteres por separado

````python
lista = list('Prueba')
lista
````

````python
type(lista)
````

Adicionalmente, hay métodos de otros tipos de variables que generan listas. Por ejemplo, el método `split()` para cadenas de caracteres genera una lista:

````python
texto = 'Curso de algoritmos y programación'
````

````python
lista = texto.split()
lista
````


# <font color='157699'> **4. Agregar, concatenar y remover elementos de listas** </font>

### <font color='46B8A9'> **Adicionar elementos a una lista** </font>  

Puede adicionarse elementos a una lista usando tres métodos diferentes.

| Función | Descripción |
|---------|-------------|
| `lista.append(elemento)` | Agrega un elemento **al final** de la lista. |
| `lista.insert(posición, elemento)` | Inserta un elemento en una posición específica de la lista. |
| `lista.extend(iterable)` | Agrega los elementos de un iterable (estructura de datos) al final de la lista. |



````python
lista = [1, 2, 3]
lista.append(8) # Agregamos el elemento 8 al final de la lista
print(lista)
````

````python
lista.insert(2, 9) # Agregamos el elemento 9 en la tercera posición
print(lista)
````

````python
lista.extend(range(11, 15)) # anexamos a la lista los números en la secuancia del 11 al 14
print(lista)
````

### <font color='46B8A9'> **Concatenación de listas** </font>  

Las listas pueden concatenarse directamente con el operador `+`
>  `lista_r = lista_1` + `lista_2`

La concatenación se realiza según el orden en el cual se escriben las listas

````python
lista1 = [1, 2, "pato"]
lista2 = [[1,3], ['a', 'b'], 6]
lista3 = lista1 + lista2
lista3
````

### <font color='46B8A9'> **Repetición** </font>  

Es posible repetir los elementos de una lista haciendo uso del operador `*`

````python
lista1 = [1, 2, 3]
lista2 = lista1 * 3
lista2
````

### <font color='46B8A9'> **Remover y eliminar elementos** </font>

Algunos de los métodos para eliminar elementos de una lista son los siguientes:

| Función | Descripción |
|---------|-------------|
| `lista.remove(elemento)` | Elimina la primera aparición de `elemento` en la lista |
| `lista.pop([i])` | Elimina y devuelve el elemento en la posición `i` de la lista; si no se proporciona `i`, se elimina y devuelve el último elemento |
| `lista.clear()` | Elimina todos los elementos de la lista |



````python
lista = [1, 8, 3, 6, 8, 9, 3]
r = lista.pop() # obtenemos el último elemento de la lista y lo removemos
print("El elemento eliminado es ", r, " la lista actualizada es ", lista )
````

````python
lista.remove(8) # remueve la primera aparición del numero 8
print(lista)
````

````python
lista.pop(1) # obtiene y remueve el elemento en la posición 1
print(lista)
````

````python
lista.clear() # Elimina todos los elementos
print(lista)
````


# <font color='157699'> **5. Funciones y métodos sobre los elementos de una lista** </font>

Algunas de las funciones que pueden aplicarse sobre los elementos de una lista son

| Función | Descripción |
|---------|-------------|
| `len(lista)` | Devuelve la cantidad de elementos de la lista |
| `max(lista)` | Devuelve el valor máximo de la lista |
| `min(lista)` | Devuelve el valor mínimo de la lista |
| `sum(lista)` | Devuelve la suma de los elementos de la lista |
| `lista.index(elemento)` | Devuelve el índice de la primera aparición de `elemento` en la lista |
| `lista.count(elemento)` | Devuelve la cantidad de veces que `elemento` aparece en la lista |
| `sorted(lista)` | Devuelve una copia ordenada de la lista |
| `reversed(lista)` | Devuelve una copia invertida de la lista |
| `lista.sort()` | Ordena los elementos de la lista en orden ascendente. |
| `lista.reverse()` | Invierte el orden de los elementos en la lista. |
| `lista.copy()` | Devuelve una copia de la lista. |



Note que estas funciones retornan una nueva variable o en algunos casos una nueva lista

### <font color='46B8A9'> **Valor mínimo** </font>  


La función `min()` permite hallar el valor mínimo en éstas.

````python
lista = [3, 1.2, 6, 5]
min(lista)
````

Note que es posible aplicarlo a distintos tipos de datos, en cuyo caso la interpretación del resultado es diferente. Por ejemplo, al aplicarla a una lista de caracteres, retorna el elemento que estaría en menor orden lexicográfico:

````python
lista = ["Pato", "Patas", "Pata"]
min(lista)
````

### <font color='46B8A9'> **Valor máximo** </font>

````python
lista = [3, 1.2, 6, 5]
max(lista)
````

De forma similar a la función `min()`, la función `max()` puede aplicarse a distintos tipos de datos, en cuyo caso la interpretación de su resultado es diferente. Por ejemplo, al aplicarla a una lista de caracteres, retorna el carácter que estaría en mayor orden lexicográfico:

````python
lista = ["Pato", "Patas", "Pata"]
max(lista)
````

### <font color='46B8A9'> **Ordenar elementos `sorted()`** </font>  


La función  `sorted()` retorna una nueva lista con los elementos ordenados

````python
lista1 = [5, 2, 1, 3, 6]
lista2 = sorted(lista1)
print("original:", lista1, "ordenada:", lista2)
````

Note que la función retorna una nueva lista mientras que la lista original no se modifica. Sin embargo, es posible aplicar el método `sort()`, el cual ordena la lista original directamente. Es decir, **no crea una nueva lista, sino que modifica la existente**

````python
lista = [5, 2, 1, 3, 6]
print(lista)
lista.sort()
print(lista)
````

Note que pueden ordenarse listas que contienen distintos tipos de datos, por ejemplo cadenas de caracteres:

````python
lista = ["Brigitte", "Andrés", "Carlos"]
lista.sort()
print(lista)
````

No es posible ordenar listas heterogéneas, es decir, con elementos de diferentes tipos de datos:

````python
lista = [3, -1, 'partida', 3+4j, 7.8]
lista.sort()
print(lista)
````

### <font color='46B8A9'> **Reversar lista `reversed()`** </font>  

La función  `reversed()` retorna los indices de la lista con los elementos en orden inverso. Siendo así, la sintaxis para generar una nueva lista con elementos reversados sería:


> ```python
nueva_lista = list(reversed(lista))
```



````python
lista1 = [5, 2, 1, 3, 6]
lista2 = reversed(lista1)
print(lista2)
print("original:", lista1, "orden reverso:", list(lista2))
````

Similarmente, se define el método `reverse()`, el cual invierte el orden de la lista original directamente. Es decir, **no crea una nueva lista, sino que modifica la existente**

````python
lista = ["Brigitte", "Andres", "Carlos"]
lista.reverse()
print(lista)
````
