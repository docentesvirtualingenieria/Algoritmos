La documentación de clases, métodos, funciones, módulos o submódulos se realiza en Python por medio de cadenas de documentación o en inglés *Docstring*.

El ***PEP 257*** documenta la semántica y las convenciones asociadas con los Docstring de Python. El objetivo de este PEP es estandarizar la estructura de alto nivel de las cadenas de documentación: qué deben contener y cómo decirlo, en forma de convenciones, no como normas o reglas de sintaxis.

Seguir estas convenciones permitirá obtener mejores resultados y compatibilidad con otros programas que administran automáticamente la documentación de scripts de Python. Aunque no hay que alarmarse, no seguir estas convenciones producirá, a lo sumo, una documentación poco organizada que no se verá tan bien.



## <font color='157699'>**¿Qué es un Docstring?**</font>



En Python las funciones y objetos cuentan con una variable especial llamada `doc`. Esta variable sirve para describir cómo se usa la función u objeto. Estas variables reciben el nombre de docstrings, cadenas de documentación.

El docstring es una cadena de caracteres que aparece como la primera línea que se escribe en la definición de clases, métodos, funciones, módulos o submódulos. Tal docstring se convierte en el atributo especial `__doc__` de ese objeto.



````python
def hola(arg):
    """Este es el docstring de la función"""
    print("Hola", arg, "!")

hola("Estudiante")
````

## <font color='056938'>Tipos de </font><font color='8EC044'>**Docstring**</font>

### <font color='056938'> Docstring de una línea</font>

Se usan para casos en los que se pueda expresar claramente en una línea las específicaciones de un objeto. Por ejemplo:


```
def function(a, b):
    """junta los objetos a y b retornádolos en una lista."""
```

### <font color='056938'>Sugerencias</font>

* Se utilizan comillas triples aun si el texto está en una sola línea. Esto facilita su posterior expansión.
* Las comillas de cierre están en la misma línea que las comillas de apertura.Esto se ve mejor para las frases de una sola línea.
* No hay espacios en blanco ni antes ni después del Dostring.
* El docstring es una frase que termina en un punto.


Por ejemplo, evite:

```
def funcion(a, b):
    """funcion(a, b) -> lista"""
```

La forma aconsejada para este docstring sería algo como:


```
def funcion(a, b):
    """Ejecuta una operación y retorna una lista."""
```

### <font color='056938'>Docstring de </font> <font color='8EC044'>**varias líneas**</font>

Se componen de una primera línea de resumen, una línea en blanco, y una descripción más elaborada de lo que realiza la función o método y posteriormente una linea en blanco donde inicia el algoritmo.


* Todo el texto de la documentación debe tener la misma sangría que las comillas en su primera línea, por ejemplo:



````python
def imaginary(real=0.0, imag=0.0):
    """
    Form a complex number.

    Parameters
    ----------
    real :float
          The real part of the number. The default is 0.
    imag : float
        The imaginary part of the number. The default is 0.

    Returns
    -------
    complex
        Return a complex number.
    """
    return complex(real,imag)

help(imaginary)
````

La documentación de un módulo generalmente debe enumerar las clases, excepciones y funciones (y cualquier otro objeto) que exporta el módulo, con un resumen de una línea de cada una. (Estos resúmenes generalmente brindan menos detalles que la línea de resumen en la cadena de documentación del objeto).

Para acceder a la documentación del código se utiliza `help` de la siguiente manera:

````python
class Clase:
    """ Este es el docstring de la clase"""
    def __init__(self):
        """Este es el docstring del inicializador de clase"""
    def metodo(self):
        """Este es el docstring del metodo de clase"""

o = Clase()
help(o)
````

la función `help` devuelve la información del documentada en la función, método o librería.

El Docstring de una función o método debe resumir su comportamiento y documentar sus argumentos, valores devueltos, efectos secundarios, excepciones planteadas y restricciones sobre cuándo se puede llamar, en el caso que corresponda. Igualmente deben indicarse los argumentos opcionales.

````python
def despedir():
    """Este es el docstring de la función despedir"""
    print("Adiós! Me despido desde la función despedir() del módulo prueba")
def saludar():
    """Este es el docstring de la función saludar"""
    print("Hola! Te saludo desde la función saludar() del módulo prueba")

help(despedir)
help(saludar)
````

Las librerías de Python generalmente vienen documentadas. El siguiente ejemplo importa la función randint de la librería random y utiliza la `help` para mostrar lo que hace la librería.

````python
from random import randint

help(randint)
````

## <font color='056938'> Forma automatica inicio de un  </font><font color='8EC044'> Docstring en Spyder</font>

Spyder es un entorno de programación que genera automáticamente el inicio de formato de documentación de una función o método.

Para generar la documentación en Spyder, primero se debe crear la función definiendo las entradas y las salidas. El siguiente ejemplo contiene una función sin documetnación:


````python
def sumapropia(a: float, b: float, c: float) -> float:

    a=5
    b=7
    c=a+b
    return c
````
