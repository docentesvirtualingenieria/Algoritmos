## <font color='8EC044'> **Ejercicio 1** </font>

Cree una función que genere  una contraseña aleatoriamente compuesta por vocales y números. Para ello, deberá recibir del usuario dos valores:  cantidad de vocales que debe contener (m) y cantidad de números (del 0 al 9) que debe contener (n). 

Por ejemplo, si `m = 3` y `n = 4`, una de las posibles contrseñas generadas por la función sería:
> `a23i6u5`

**Recomendación:** Use el método `sample` de la librería `random`.

````python
# Inserte aquí su respuesta
import random 

def clave(m, n):
    vocales = random.sample('aeiou', m, counts=None)
    numeros = random.sample('0123456789', n, counts=None)
    res1 = random.sample(vocales + numeros, m+n, counts=None)
    resultado = ''
    for i in range(len(res1)):
        resultado += res1[i]
    return resultado
````

````python
# Pruebe la función creada aquí
clave(3,4)
````

# <font color='157699'> **7. Indexación por intervalos: slicing** </font>

- La indexación se puede hacer por grupos, para extraer partes de una cadena de caracteres. Esto se conoce como "*slicing*", que en inglés significa rebanado.
- La sintaxis requiere del uso de los dos puntos (`:`) para separar el inicio, el final, y el incremento en los índices:
`variable[i_ini:i_fin:incr]`
- El incremento siempre debe ser un número entero.
- Si el incremento es la unidad, se puede omitir, y usar únicamente inicio y fin: `variable[i_ini:i_fin]`
- Si se desea iniciar desde el primer carácter, no hace falta escribirlo: `variable[:i_fin]`
- Cuando se desea indexar hasta el último carácter, tampoco hace falta escribirlo: `variable[i_ini:]`

La indexación por intervalos o "*slicing*" es similar a la función `range()`, ya que al ejecutar la acción se incluye el valor de inicio, pero no el valor final. Es decir, que se llega hasta un incremento antes del valor final (si no se especifica el incremento, éste es 1 de forma predeterminada).

````python
nombre
````

````python
nombre[1:5]   
````

````python
nombre[:3]
````

````python
nombre[3:]
````

````python
nombre[2:5]
````

La indexación por intervalos también funciona con índices negativos:

````python
nombre
````

````python
nombre[-3:]
````

````python
nombre_completo
````

````python
nombre_completo[-5:]
````

Cuando no se incluye índice de inicio ni de final, se toma la cadena de caracteres completa:

````python
nombre[:]
````

Se puede usar un incremento (**siempre entero**) para acceder a los caracteres "a saltos":

````python
nombre[::2]
````

````python
nombre[1::2]
````

Usando un incremento negativo, se puede indexar en dirección de izquierda a derecha. Por ejemplo, esto se puede usar para invertir una cadena de caracteres:

````python
nombre[::-1]
````

Esta funcionalidad permite reconocer palíndromos de forma sencilla en Python:

````python
palindromo = 'reconocer'
palindromo == palindromo[::-1]
````

**Resumen:** <br> sea `s = 'abcdefghij'` <br>

| Código | Resultado | Descripción|
| :--- | :--- | :--- |
|`s[2:5]` |`cde` | Caracteres en los índices 2 a 4 |
|`s[:5]` |`abcde` | Primeros cinco caracteres (0 a 4) |
|`s[5:]` | `fghij` | Caracteres desde el índice 5 hasta el final |
|`s[-2:]` | `ij` | Dos últimos caracteres |

|`s[1:7:2]` | `bdf` | Caracteres del índice 1 al 6, cada dos posiciones |
|`s[::-1]` | `jihgfedcba` | Un paso negativo invierte la cadena de caracteres |

En este ejercicio se debe crear una versión muy básica de encriptación. Para ello, crearemos una función que recibe una cadena de caracteres y retorna otra cadena en la cual todas las vocales han sido rempazadas por su [código ASCII](https://concepto.de/wp-content/uploads/2018/04/Codigo-ASCII-min-e1523384955302.jpg) equivalente. 

Por ejemplo, al recibir el texto:

> `Hola mundo`

La función debe retornar:

> `H111l97 m117nd111`


Tenga en cuenta las operaciones de indexación y la inmutabilidad de las cadenas de caracteres.

**Recomendación:** Use la fución `ord()`, la cual retorna el código ASCII del carácter que se le dé como argumento.

````python
# Inserte aquí su respuesta
def encript_voc(palabra):
    resultado = ''
    vocales = 'aeiouAEIOUáéíóúÁÉÍÓÚ'
    for i in range(len(palabra)):
        if palabra[i] in vocales:
            resultado += str(ord(palabra[i]))
        else:
            resultado += palabra[i]
    return resultado
````

# <font color='157699'>**11. Iteración o recorrido de cadenas de caracteres**</font>

Una cadena de caracteres puede recorrerse usando la estructura de repetición `for` que vimos anteriormente.

````python
curso = 'Programación'
for char in curso:
    print(char)
````

 Una segunda forma es llevando el  control de la posición de cada carácter a través de una variable que usamos con indexación:

````python
print('Índice', 'Carácter', sep='\t')
for i in range(len(string_1)):
    print(i, string_1[i], sep='\t')
````
