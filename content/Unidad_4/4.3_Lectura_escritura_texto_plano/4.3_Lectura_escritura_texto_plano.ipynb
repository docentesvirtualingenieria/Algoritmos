## <font color='157699'> **Lectura y escritura de archivos de texto** </font>

Una forma de leer o escribir un archivo en Python es usar la función  <code>open</code>. La función <code>open</code> proporciona un objeto tipo archivo que contiene los métodos y atributos que necesita para leer, guardar y manipular dicho archivo. En este notebook, solo se cubrirán archivos de texto plano. El primer parámetro que necesita es una cadena de caracteres con la ruta del archivo incluyendo el nombre del mismo. El segundo parámetro se describe a continuación:


https://docs.python.org/es/3/library/functions.html#open

| Carácter | Significado                                                           |
| -------- | --------------------------------------------------------------------- |
| `’r’`    | abierto para lectura (por defecto)                                    |
| `'w'`    | abierto para escritura, truncando primero el fichero                  |
| `'x'`    | abierto para creación en exclusiva, falla si el fichero ya existe     |
| `’a’`    | abierto para escritura, añadiendo al final del fichero si este existe |
| `’+’`    | abierto para actualizar (lectura y escritura)                         |

### <font color='46B8A9'> **Creación de un archivo de trabajo** </font>


````python
"""
Ejecutar y no modificar.
Creación de un archivo de texto plano para trabajar sobre este.  
"""

# librería par trabajar con el sistema operativo
import os 

directorio = "/content/"

archivo = "nombres.txt" 

ruta = os.path.join(directorio, archivo)

with open(ruta, 'w') as archivo_txt:
    archivo_txt.write("Carlos\n")
    archivo_txt.write("Pedro\n")
    print("Creación del archivo exitosa")
````

Al ejecutar el código, se crea un archivo el cual se encuentra alojado en el disco duro de la maquina virtual, para el caso de Google Colab.

### <font color='46B8A9'> **Lectura de un archivo** </font>
A continuación se presenta una forma de abrir el archivo, ver algunos atributos e imprimir el contenido de archivo de texto plano. 

````python
ruta = "/content/nombres.txt"
archivo_txt = open(ruta, "r")
````

````python
"""
Volver a abir el archivo, imprimir todo su contenido usando 
un ciclo y finalmente cerrar el archivo
"""

archivo_txt = open(ruta, "r")

for linea in archivo_txt:
    print(linea)

archivo_txt.close()
````

### <font color='46B8A9'> **Una mejor manera de abrir un archivo** </font>

Usar la declaración <code>with</code> es una mejor práctica, cierra automáticamente el archivo incluso si el código encuentra una excepción o error. El código ejecutará todo en el bloque y luego cerrará el archivo.

````python
# Leer el archico completo

with open(ruta, "r") as archivo_txt:
    print(archivo_txt.read())
````

````python
# Leer los primeros cuatro caracteres

with open(ruta, "r") as archivo_txt:
    print(archivo_txt.read(4))
````

Una vez que se llama al método <code>.read(4)</code>, se llaman los primeros 4 caracteres. Si volvemos a llamar al método, se llamarán los siguientes 4 caracteres. El resultado de la siguiente celda demostrará el proceso para diferentes entradas al método <code>read()</code>:

````python
with open(ruta, "r") as archivo_txt:
    print(archivo_txt.read(4))
    print(archivo_txt.read(4))
    print(archivo_txt.read(4))
````

Leer la primer línea del archivo usando el método <code>readline()</code>:

````python
# Leer una línea

with open(ruta, "r") as archivo_txt:
    print(archivo_txt.readline())
````

Leer línea a línea el archivo usando el método <code>readline()</code>:

````python
# Leer línea a línea

with open(ruta, "r") as archivo_txt:
    print(archivo_txt.readline())
    print(archivo_txt.readline())
    print(archivo_txt.readline())
````

Es posible usar un ciclo para iterar a través de cada línea:

````python
with open(ruta,"r") as archivo_txt:
        i = 0;
        for linea in archivo_txt:
            print("Línea", str(i), ": ", linea[:-1])
            i = i + 1;
````

Es posible usar el método <code>.readlines()</code> para guardar el archivo de texto en una lista, dónde cada línea del archivo de texto, es un elmento de la lista:

````python
# Leer todas las líneas y guarda el contenido en una lista 

with open(ruta, "r") as archivo_txt:
    archivo_lista = archivo_txt.readlines()
    print(archivo_lista)
````

## <font color='157699'> **Escritura de archivos de texto** </font>

Para escribir en un archivo, se debe abrir en modo escritura. Para esto se debe pasar el segundo argumento como: <code>'w'</code>. Luego con el  método <code>.write()</code> para agregar texto al archivo. 

````python
# Primero mostramos el contenido actual del arhivo

with open(ruta, "r") as archivo_txt:
    for linea in archivo_txt:
        print(linea[:-1])
````

````python
# Escribir en el archivo

with open(ruta, "w") as archivo_txt:
    archivo_txt.write('Juan\n')
    archivo_txt.write('Ana\n')
````

````python
# Mostramos el contenido actualizado del arhivo

with open(ruta, "r") as archivo_txt:
    print(archivo_txt.read())
````

Para evitar reemplazar el contenido original del archivo, se debe abrir en modo para agregar <code>'a'</code>. De esta forma puede agregar una nueva línea al final del contenido del archivo. 

````python
# Escribir nuevas líneas en el archivo de texto

with open(ruta, "a") as archivo_txt:
    archivo_txt.write('Clara\n')
    archivo_txt.write('Julia\n')
````

````python
# Mostramos el contenido actualizado del arhivo

with open(ruta, "r") as archivo_txt:
    print(archivo_txt.read())
````

También es posible reemplazar el contenido del archivo abriendo en el modo <code>'r+'</code>.

````python
# Escribir nuevas líneas desde el inicio 
with open(ruta, "r+") as archivo_txt:
    archivo_txt.write('Manuel\n')
    
````

````python
# Mostramos el contenido actualizado del arhivo

with open(ruta, "r") as archivo_txt:
    print(archivo_txt.read())
````

## <font color='157699'> **Copiar un archivo** </font>

Se pueden utilizar los conceptos vistos para realizar la copia de un archivo. 


````python
# Copiar archivo a otro

with open("/content/nombres.txt",'r') as archivo_txt:
    with open("/content/nombres_2.txt",'x') as nuevo_txt:
          for linea in archivo_txt:
                nuevo_txt.write(linea)
````

````python
# Mostramos el contenido del nuevo archivo 

with open("/content/nombres_2.txt", "r") as nuevo_txt:
    print(nuevo_txt.read())
````

En este caso se creo un nuevo archivo llamado <code>'nombres_2.txt'</code> en modo <code>'x'</code>. Lo que significa, creación exclusiva. Si se ejecuta nuevamente saldrá un error. 

````python
# Copiar archivo a otro

with open("/content/nombres.txt",'r') as archivo_txt:
    with open("/content/nombres_2.txt",'x') as nuevo_txt:
          for linea in archivo_txt:
                nuevo_txt.write(linea)
````
