# <font color='157699'> **4. Recorrido de listas** </font>

Las listas pueden recorrerse con ciclos `for` a tráves del comando `in`, el cual recorre todos los elementos de la lista en el orden en que están almacenados, pero sin necesidad de referirse a su índice.

````python
lista = [1, 2, "a", 6]
for elemento in lista:
    print(elemento)
````

````python
print('Elemento', 'Tipo', sep='\t')
print('-'*30)
for item in lista:
    print(item, type(item), sep='\t\t')
````

**Ejemplo: Identificación del tipo de elementos en una lista**

````python
lista = ['o', 5, 7.2, 'palabra', 10]
ent = 0 # cantidad de enteros
cad = 0 # cantidad de cadenas de caracteres
real = 0 # cantidad de números reales
for item in lista:
    if type(item) == int:
        ent += 1
    elif type(item) == str:
        cad += 1
    elif type(item) == float:
        real += 1
print('\nLa lista tiene %d elementos que son números enteros' % ent)
print('\nLa lista tiene %d elementos que son números reales' % real)
print('\nLa lista tiene %d elementos que son cadenas de caracteres' % cad)
````

Las listas también se pueden recorrer haciendo referencia explícita al índice del elemento. En este caso es necesario indicar la secuencia de indices, para lo que usamos la función `len()`, con la cual se obtiene la cantidad de elementos, así como la función `range()`, con al cual se crea la secuencia de índices.

````python
lista = [1, 2, "a", 6]
for i in range(len(lista)):
    print(lista[i])
````

### <font color='46B8A9'> **Adicionar elementos a una lista** </font>  

Puede adicionarse elementos a una lista usando tres métodos diferentes.

| Función | Descripción |
|---------|-------------|
| `lista.append(elemento)` | Agrega un elemento **al final** de la lista. |
| `lista.insert(posición, elemento)` | Inserta un elemento en una posición específica de la lista. |
| `lista.extend(iterable)` | Agrega los elementos de un iterable (estructura de datos) al final de la lista. |



````python
lista = [1, 2, 3]
lista.append(8) # Agregamos el elemento 8 al final de la lista
print(lista)
````

````python
lista.insert(2, 9) # Agregamos el elemento 9 en la tercera posición
print(lista)
````

````python
lista.extend(range(11, 15)) # anexamos a la lista los números en la secuancia del 11 al 14
print(lista)
````

### <font color='46B8A9'> **Concatenación de listas** </font>  

Las listas pueden concatenarse directamente con el operador `+`
>  `lista_r = lista_1` + `lista_2`

La concatenación se realiza según el orden en el cual se escriben las listas

````python
lista1 = [1, 2, "pato"]
lista2 = [[1,3], ['a', 'b'], 6]
lista3 = lista1 + lista2
lista3
````

### <font color='46B8A9'> **Repetición** </font>  

Es posible repetir los elementos de una lista haciendo uso del operador `*`

### <font color='46B8A9'> **Remover y eliminar elementos** </font>

Algunos de los métodos para eliminar elementos de una lista son los siguientes:

| Función | Descripción |
|---------|-------------|
| `lista.remove(elemento)` | Elimina la primera aparición de `elemento` en la lista |
| `lista.pop([i])` | Elimina y devuelve el elemento en la posición `i` de la lista; si no se proporciona `i`, se elimina y devuelve el último elemento |
| `lista.clear()` | Elimina todos los elementos de la lista |



````python
lista = [1, 8, 3, 6, 8, 9, 3]
r = lista.pop() # obtenemos el último elemento de la lista y lo removemos
print("El elemento eliminado es ", r, " la lista actualizada es ", lista )
````

````python
lista.remove(8) # remueve la primera aparición del numero 8
print(lista)
````

````python
lista.pop(1) # obtiene y remueve el elemento en la posición 1
print(lista)
````

````python
lista.clear() # Elimina todos los elementos
print(lista)
````

### <font color='46B8A9'> **Ordenar elementos `sorted()`** </font>  


La función  `sorted()` retorna una nueva lista con los elementos ordenados

````python
lista1 = [5, 2, 1, 3, 6]
lista2 = sorted(lista1)
print("original:", lista1, "ordenada:", lista2)
````

Note que la función retorna una nueva lista mientras que la lista original no se modifica. Sin embargo, es posible aplicar el método `sort()`, el cual ordena la lista original directamente. Es decir, **no crea una nueva lista, sino que modifica la existente**

````python
lista = [5, 2, 1, 3, 6]
print(lista)
lista.sort()
print(lista)
````

Note que pueden ordenarse listas que contienen distintos tipos de datos, por ejemplo cadenas de caracteres:

````python
lista = ["Brigitte", "Andrés", "Carlos"]
lista.sort()
print(lista)
````

No es posible ordenar listas heterogéneas, es decir, con elementos de diferentes tipos de datos:

````python
lista = [3, -1, 'partida', 3+4j, 7.8]
lista.sort()
print(lista)
````

### <font color='46B8A9'> **Reversar lista `reversed()`** </font>  

La función  `reversed()` retorna los indices de la lista con los elementos en orden inverso. Siendo así, la sintaxis para generar una nueva lista con elementos reversados sería:


> ```python
nueva_lista = list(reversed(lista))
```



````python
lista1 = [5, 2, 1, 3, 6]
lista2 = reversed(lista1)
print(lista2)
print("original:", lista1, "orden reverso:", list(lista2))
````

Similarmente, se define el método `reverse()`, el cual invierte el orden de la lista original directamente. Es decir, **no crea una nueva lista, sino que modifica la existente**

````python
lista = ["Brigitte", "Andres", "Carlos"]
lista.reverse()
print(lista)
````

# <font color='157699'> **7. Programación compacta con listas <font color='46b8a9'> (*list comprehensions*)** </font>

Consiste en definir una o varias operaciones, incluidas la evaluación de condiciones y la estructuras de repetición, directamente en una línea de código entre corchetes:
    
```python
[f(var) for var in iterable if var > 0]
```
    
Donde `f()` es una función que genera el valor de cada elemento de la lista. `var` es una variable transitoria que toma los valores de los elementos en la estructura de datos `iterable`, si se cumple la condición dada.
    
La programación compacta con listas ahorra mucho código, y permite crear resultados útiles de forma rápida y sencilla. 
    
Pro ejemplo, sada una lista de elementos `L`, estamos interesados en crear una nueva lista con cada uno de sus elementos elevados al cuadrado, sólo si éstos son mayores que 4. 

La forma convecional, usando ciclos `for`, sería:  

````python
lista1 = [1, 2, 3, 4, 5, 6]
lista2 = []
for item in lista1:
    if item > 4:
        lista2.append(item**2)
lista2
````

La forma compacta usado `list comprehensions` sería (Se usa una línea de código, en vez de 4):

````python
lista1 = [1, 2, 3, 4, 5, 6]
lista2 = [item**2 for item in lista1 if item > 4]
lista2
````
